## Как обучается нейросеть?

В процессе обучение параметры сети (веса нейронов) настраиваются с использованием градиентного спуска, минимизируя некоторую функцию потерь.

Для ускорения используется механизм **обратного распространения ошибки**

image.png

### Обратное распространение ошибки

- Обратное распространение направлено на минимизацию функции потерь путем корректировки весов и смещений сети. 
- Степень корректировки определяется градиентами функции потерь по отношению к этим параметрам. (Градиент - вектор частных производных функции потерь по весам нейронной сети)
- Производная функции С отражает чувствительность к изменению значения функции (выходного значения) относительно изменения ее аргумента х (входного значения). 

  Градиент показывает, насколько необходимо изменить параметр w (в положительную или отрицательную сторону), чтобы минимизировать С.

### Основные проблемы в обучении

- **Затухание градиента** - если среди производных для вычисления градиента многие будут близки к нулю, произведение будет околонулевым. При нулевой производной по весу *w*, он не будет обновляться во время обучения. В итоге может сложиться ситуация, когда обновляются только последние слои.

  Возможные решения: замена функции активации, `skip-connection`...

- **Взрыв градиента** - если все градиенты большие по модулю, может произойти переполнение, в таком случае веса сильно меняются, сеть теряет стабильность.

  Возможные решения: регуляризация весов, `gradient clipping` (обрезка градиентов параметров во время обратного распространения по максимальному значению или предельной норме).


**Skip connection**  позволяет градиенту 
проходить не только через нелинейное преобразование слоя, но также и через тождественное преобразование.

## Функции активации

**Функция активации**  — нелинейное преобразование, поэлементно применяющееся к пришедшим на вход данным

### Sigmoid

- на хвостах обладает практически нулевой производной, что может привести к затуханию градиента
- максимальное значение производной сигмоиды составляет 0.25 - приводит к затуханию градиента на большом количестве слоев
- требует вычисления экспоненты, что является достаточно вычислительно сложной операцией

image.png

### ReLU

Производная вычисляется просто и быстро

- необходимо доопределить в нуле
- для отрицательных значений производная равна нулю, что может привести к затуханию градиента

image.png

### Leaky ReLU

Сглаживает недостатки ReLU

- небольшой уклон слева позволяет получить более симметричную относительно нуля область значений
- меньше затухание градиента благодаря наличию ненулевого градиента и слева, и справа от нуля
